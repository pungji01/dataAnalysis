###[실습문제' while 반복문(퀴즈 맞히기)

```python
print('프로그램을 작성할 때 일정한 구간의 명령을\n반복해서 실행하도록 구성한 문장을 무엇이라고 할까요?')
while True:
    answer = input('정답: ')
    if answer == '반복문' :
        print('정답!')
        break
    print('오답')
print('프로그램 종료!')
```

```python
# continue 사용
print('프로그램을 작성할 때 일정한 구간의 명령을\n반복해서 실행하도록 구성한 문장을 무엇이라고 할까요?')
while True:
    answer = input('정답: ')
    if answer == '반복문' :
        print('정답!')
        break
    elif answer == '반복':
        print('좀 더 고민해 보세요~')
        continue
    print('오답')
print('프로그램 종료!')
```

###[실습문제] 금리 계산하기

```python
seed, target = 10000000, 20000000
balance, interest = 10000000, 0.07
year = 0
while balance <= target:
    balance += balance * interest
    year += 1
    print(f"year : {year}, balance:{round(balance):,}")
print(f"{seed:,}원이 {target:,}원이 되는데 {year}년이 걸립니다.")
```

###[실습문제] 1 ~ N까지 3의 배수 합계 구하기

```python
result = [i for i in range(1,1001)]
print(result)
result = [i for i in range(1,1001) if i%3 == 0]
print(result)
result = sum([i for i in range(1,1001) if i%3 == 0])
print(f'{result:,}')
```

###[실습문자] 숫자 입력 받아 별 출력하기

```python
star = '⭐'
numbers = input('숫자 여러 개 입력(0~9): ')

for number in numbers:
    for _ in range(int(number)*2):
        print(star, end='')
    print()
```

###[실습문제] 무인 호텔관리 시스템

```python
'''
1. 체크인
방 호수(1번부터 시작)를 입력 받습니다.
방이 이미 입실 중이면 "입실 중인 방은 체크인 하실 수 없습니다."를 출력하세요.
빈 방인 경우 "입실 완료!"를 출력하고 메뉴로 돌아갑니다.
2. 체크아웃
방 호수(1번부터 시작)를 입력 받습니다.
방이 빈 방이면 "빈 방은 체크아웃 하실 수 없습니다."를 출력하세요.
체크인 상태인 경우 "퇴실 완료!"를 출력하고 메뉴로 돌아갑니다.
3. '방호수 - 상태'를 출력합니다.
출력 예)
1호 : 빈 방
2호 : 입실 중
3호 : 입실 중
4호 : 빈 방
5호 : 빈 방
0. 종료
반복을 종료하고 '영업 종료' 를 출력합니다.
'''
```

```python
print('호텔관리 시스템입니다.')
menu = '''
1. 체크인
2. 체크아웃
3. 입실현황
0. 종료'''
number = int(input('호텔 방의 개수는(입력): '))
hotel = [0 for _ in range(number)]
print(f"방 현황: {hotel} (0:빈방, 1:입실중)")
while True:
    print('-------------------')
    print(menu)
    select = input('-->메뉴 선택: ')
    if select == '0':
        break

    if select == '1': # 체크인
        num = int(input(f'체크-in하실 방 호수(1~{len(hotel)}): '))
        if hotel[num-1]:   # 이미 입실 중일 때
            print(f"{num}번 방은 이미 입실 중입니다.")
        else:
            hotel[num-1] = 1
        continue

    if select == '2': # 체크아웃
        num = int(input(f'체크-out하실 방 호수(1~{len(hotel)}): '))
        if hotel[num-1]: # 체크아웃 처리
            hotel[num-1] = 0    # 빈 방 상태로 만듦
        else:
            print(f'{hotel[num]}번 방은 빈 방입니다. 빈 방은 체크-out하실 수 없습니다.')

    if select == '3': # 현황보기
        for idx, room in enumerate(hotel): #[0, 0, 0]
            if room == 0:
                print(f"{idx+1}호: 빈방")
            elif room == 1:
                print(f"{idx+1}호: 입실중")


print('-------------------')
print('프로그램 종료!')
```

```python
import random
dice = [0,0,0,0,0,0]
for _ in range(100):
    x = random.randint(1,6)
    dice[x-1] += 1
print(f"숫자별 개수 : {dice}")

maxValue = max(dice)
max_dice = [idx+1 for idx, v in enumerate(dice) if v == maxValue]
print(f"가장 많이 나온 숫자 : {max_dice}, 횟수 : {maxValue}")
```
